```python
# data to plot
n_data = 4
values_1 = (90, 55, 40, 65)
values_2 = (85, 62, 54, 20)

# Configuramos el plot
fig, ax = plt.subplots()
index = np.arange(n_data)
bar_width = 0.35

# Configuramos las barras del primer conjunto de valores
rects1 = plt.bar(index, values_1, bar_width,
                 color='r',
                 label='Valores 1')

# Configuramos las barras del segundo conjunto de valores
rects2 = plt.bar(index + bar_width, values_2, bar_width,
                 color='g',
                 label='Valores 2')

# Insertamos los ttulos y la leyenda del grfico
plt.xlabel('Etiquetas')
plt.ylabel('Valores')
plt.title('Ttulo del grfico')
plt.xticks(index + bar_width, ('A', 'B', 'C', 'D'))
plt.legend()

# Pintamos el grfico y lo mostramos
plt.tight_layout()
plt.show()
```
