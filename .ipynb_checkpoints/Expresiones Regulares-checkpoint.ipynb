```python
import re
```

```python
mensaje = "Esto es un mensaje de ejemplo del curso de Python"
```

```python
match = re.search("curso", mensaje)
```

```python
match.end()
```

```python
match2 = re.match("Esto", mensaje)
```

```python
match2.end()
```

```python
mensajeSinEspacios = re.split("o", mensaje)
```

```python
mensajeSinEspacios
```

```python
patron = re.findall("de", mensaje)
```

```python
patron
```
