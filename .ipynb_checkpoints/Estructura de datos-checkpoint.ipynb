```python
contactos = {
    "persona1": {"nombre": "Anthony", "telefono": "1234567", "email": "anthony@ejemplo.com"},
     "persona2": {"nombre": "Ana", "telefono": "123456789", "email": "ana@ejemplo.com"},
     "persona3": {"nombre": "Susana", "telefono": "123456789", "email": "susana@ejemplo.com"},
}

print(contactos["persona2"]["email"])
contactos["persona4"] = {"nombre": "Fredy", "telefono": "123434322", "email": "fredy@ejemplo.com"}
contactos["persona1"]["telefono"] = "7645343"

for contacto in contactos:
    print(contactos[contacto]["nombre"])

```
