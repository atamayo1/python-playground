```python
# Division con resultado  de un entero
division = 7 // 6
# multiplicaciones complejas
multiplica = 7.15 * 0.156j
# elevar x a la y
elevacion = 7 ** 6
# tener el residuo entre dos numeros
residuo = 7 % 6

print("division:", division)
print("multiplicacion:", multiplica)
print("elevacion:", elevacion)
print("residuo:", residuo)
```
