```python
from plotly.offline import init_notebook_mode

init_notebook_mode(connected=True)
```

```python
from plotly.offline import iplot
import plotly.graph_objects as go
import numpy as np
```

```python
valoresY = np.array([1,5,7,2,8])
valoresX = np.array([5,4,3,2,1])

data = [go.Scatter(x=valoresX, y=valoresY, mode='lines', line = dict(color='red', width=5, dash='dash'))]

iplot(data)
```

```python
data2 = [go.Scatter(x=valoresX, y=valoresY, mode='markers', marker = dict(color='red', symbol='pentagon', size=10))]

iplot(data2)
```

```python
valoresY2 = np.array([1,5,7,2,8])
valoresX2 = np.array([1,2,3,4,5])

data3 = [go.Scatter(x=valoresX, y=valoresY, mode='lines', line = dict(color='pink', width=15,dash='dash')), go.Scatter(x=valoresX2, y=valoresY2, mode='lines', line = dict(color='blue', width=2,dash='dash'))]

iplot(data3)
```

```python
valoresY = np.array([1,5,7,2,8])
valoresX = np.array([5,4,3,2,1])

valoresY2 = np.array([1,5,7,2,8])
valoresX2 = np.array([1,2,3,4,5])

data4 = [go.Histogram(x=valoresX, y=valoresY, marker_color='red', xbins=dict(start=2, end=4, size=0.6))]
data5 = [go.Histogram2d(x=valoresX, y=valoresY), go.Histogram(x=valoresX2, y=valoresY2)]

iplot(data4)
iplot(data5)
```

```python
valoresY = np.array([1,5,7,2,8])
valoresX = np.array([5,4,3,2,1])

valoresY2 = np.array([1,5,7,2,8])
valoresX2 = np.array([1,2,3,4,5])

data = [go.Bar(x=valoresX, y=valoresY, name="Valores 1"), go.Bar(x=valoresY2, y=valoresX2, name="Valores 2")]

iplot(data)
```

```python
valores = np.array([1,5,7,2,8])

labels = ["red", "cyan", "blue", "green", "orange"]

colors = ["red", "cyan", "blue", "green", "orange"]

data = [go.Pie(values=valores, labels=labels, marker=dict(colors=colors), textinfo='label+percent')]

iplot(data)
```
